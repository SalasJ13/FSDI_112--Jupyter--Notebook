```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        if not self.head:
            self.head = Node(item)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
            
    def insert(self, index, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                temp_node = self.head
                self.head = new_node
                new_node.next = temp_node
            else:
                counter = 0
                prev_node = None
                current = self.head
                while current.next and index != counter:
                    counter += 1
                    prev_node = current
                    current = current.next
                if not current.next:
                    current.next = new_node
                else:
                    prev_node.next = new_node
                    new_node.next = current
        
    def remove(self, value):
        current = self.head
        if current.data == value:   
            self.head = current.next
        else:
            prev_node = None
            while current.next and current.data != value: 
                prev_node = current
                current = current.next
            if current.data == value:
                prev_node.next = current.next
            else:
                raise ValueError("%s not is list" % repr(value))
                
                
    def __len__(self):
        counter = 0
        current = self.head
        while current:
            counter += 1
            current = current.next
        return counter
    
    def __getitem__(self, index):
        counter = 0
        current = self.head
        while current and index != counter:
            counter += 1
            current = current.next
        if not current:
            raise IndexError("list index not in range")
        return current.data
    
        
        
    def __str__(self):
        out_str = "["
        current = self.head
        while current.next:
            out_str += "%s, " % repr(current.data)
            current = current.next
        out_str += "%s]" % str(current.data)
        return out_str
```
